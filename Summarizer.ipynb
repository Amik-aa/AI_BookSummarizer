```
import fitz
import re

def extract_clean_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""

    for page in doc:
        page_text = page.get_text("text")
        text += page_text.strip() + "\n"

    # Replace multiple spaces with a single space
    text = re.sub(r'\s+', ' ', text)

    return text.strip()


pdf_path = "/content/Ng_MachineLearningYearning.pdf"
book_text = extract_clean_text_from_pdf(pdf_path)

# Display first 500 characters
print(book_text[:500])

```

```
from transformers import pipeline

def summarize_text(text, max_length=300):
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    summary = summarizer(text, max_length=max_length, min_length=100, do_sample=False)
    return summary[0]['summary_text']


summary_text = summarize_text(book_text[:2000])
print(summary_text)
```

```
from transformers import AutoTokenizer
from tqdm import tqdm

# Initialize the tokenizer (using GPT-2 for splitting)
tokenizer = AutoTokenizer.from_pretrained("gpt2")

def chunk_text(text, chunk_size=2000):
    tokens = tokenizer.encode(text)
    chunks = [tokens[i : i + chunk_size] for i in range(0, len(tokens), chunk_size)]
    return [tokenizer.decode(chunk) for chunk in chunks]

# Sample book text 
book_text = "This is a sample book text. It contains multiple sentences. The function should split it into chunks."

# Splitting the extracted book text into chunks
text_chunks = chunk_text(book_text)


def summarize_text(text):
    return text[:100] + "..."  # Simulating a summary by truncating the text

# Summarizing each chunk
summaries = [summarize_text(chunk) for chunk in tqdm(text_chunks)]

# Combining summaries into a final concise version
final_summary = " ".join(summaries)

print(final_summary)
```
